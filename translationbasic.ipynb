```python
import pandas as pd
# Create a dataset (same concept, but original)
data = pd.DataFrame({
    "Vegetable Names (Hindi)": [
        "गाजर",
        "शिमला मिर्च",
        "भिन्डी",
        "मक्का",
        "लाल मिर्च",
        "खीरा",
        "कढ़ी पत्ता",
        "बैंगन",
        "लहसुन",
        "अदरक"
    ]
})
data.to_csv("hindi_vegetables.csv", index=False)

data
```

```python
hindi_to_english = {
    "गाजर": "carrot",
    "शिमला मिर्च": "capsicum",
    "भिन्डी": "ladyfinger",
    "मक्का": "corn",
    "लाल मिर्च": "red chili",
    "खीरा": "cucumber",
    "कढ़ी पत्ता": "curry leaves",
    "बैंगन": "brinjal",
    "लहसुन": "garlic",
    "अदरक": "ginger"
}
```

```python
# Load dataset
df = pd.read_csv("hindi_vegetables.csv")
df["Vegetable Names (English)"] = df["Vegetable Names (Hindi)"].map(hindi_to_english)

df
```

```python
for hindi, english in hindi_to_english.items():
    print(f"{hindi}  →  {english}")
```
